```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(z) :  
  g = np.zeros(np.shape(z));
  g = 1/(1+np.exp(-z));
  return g
```

```python
def normalize(X):  
  mu=np.mean(X);
  X_norm=(X-mu);
  #sigma=np.std(X)+0.1;
  ran=np.max(X)-np.min(X);
  X_norm=X/(ran+0.00001);
  return X_norm;
```

```python
def costfunction(theta,X, y, lamda):
  m, n = np.shape(X);
  hx=sigmoid(X@theta);
 # reg=np.sum((np.square(theta)))-theta(0)^2;
  J = np.sum(-y.T@np.log(hx)-(1-y.T)@np.log(1-hx))/m + (lamda/2*m)*(np.sum((np.square(theta)))-theta[0]**2);
  return J
```

```python
def grad_des(X,y1,theta,alpha,itter):
  m,n=np.shape(X)
  for i in range(0,itter):
    for j in range(0,10):
       hx=sigmoid(X@theta[:,j])
       print("For iteration:", i, " Value of hx is:",hx, "for class: ",j)
       for k in range(0,n):  
          theta[k,j]-=(alpha/m)*np.sum(hx-y1[:,j]*X[:,k])
  #print('Shape of theta at the end of grad_desc is', np.shape(theta));
  return theta
```

```python
def predict(theta,X):
  #print('shape of theta in predict is: ', np.shape(theta))
  m, n = np.shape(X);
  v=sigmoid(X@theta)
  ind=np.zeros([m,1])
  ind=np.argmax(v, axis=1, out=None);
  return ind
```

```python
train_df = pd.read_csv("/content/sample_data/mnist_train_small.csv",header=None)
test_df  = pd.read_csv("/content/sample_data/mnist_test.csv",header=None)
data=(train_df[:]).T;
y=(data[:1]).T
X=normalize((data[1:]).T)
m, n = np.shape(X)
O= np.array([np.ones(m)]).T
X= np.hstack([O,X])
c=10;
y=np.array(y)
lamda=0.001;
y1=np.zeros([m,c])
for i in range(0, c):
  for j in range(0,m):
    k=y[j].copy()
    if k.item() == i:
      y1[j,i]=1
    else:
      y1[j,i]=0

theta = np.zeros([n+1,c])
#print(np.shape(X)) #(20000,785)
#print(np.shape(y1)) #(20000,10)
#print(np.shape(theta)) #(785,10)
#print(np.shape(y)) #(20000,1)

theta = grad_des(X,y1,theta,0.002,100)
#Now let's work on predicting
data1=(test_df[:]).T;
Xp=normalize((data1[1:]).T)
yp=(data1[:1]).T
yp=np.array(yp)
m2, n2 = np.shape(Xp)
Op= np.array([np.ones(m2)]).T
Xp= np.hstack([Op,Xp])
pred=np.zeros([m2,1])
pred = predict(theta, Xp);
print('\nTraining Set Accuracy:')
s=0
#print(np.shape(pred))
#print(np.shape(yp))
for i in range(0,m2):
  d=pred[i];
  d1=yp[i];
  if d.item()==d1.item():
    s=s+1
print((s/m2) * 100);
```
